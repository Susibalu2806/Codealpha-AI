```python
import pandas as pd
import gradio as gr

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

print("All libraries imported successfully!")
```

```python
faq_data = {
    "Question": [
        "What are the college timings?",
        "Where is the college library?",
        "How can I pay my college fees?",
        "When do semester exams start?",
        "How can I apply for a bonafide certificate?",
        "Where is the principal's office?",
        "Is there a hostel facility?",
        "How can I contact the college office?",
        "When does the admission process begin?",
        "Does the college provide placement training?",
        "What courses are offered in the college?",
        "Is there a college bus facility?",
        "How can I access the college Wi-Fi?",
        "What are the library timings?",
        "Can I participate in extracurricular activities?"
    ],

    "Answer": [
        "The college timings are from 8:30 AM to 4:30 PM.",
        "The library is located on the first floor of the main academic block.",
        "You can pay your college fees through the college office or the online student portal.",
        "Semester exams are conducted at the end of each semester as per the academic calendar.",
        "You can apply for a bonafide certificate through the administrative office.",
        "The principal's office is located in the administration block.",
        "Yes, hostel facilities are available for both boys and girls.",
        "You can contact the college office by phone or email during working hours.",
        "Admissions usually begin during the months of May and June.",
        "Yes, the college provides placement training and career guidance.",
        "The college offers undergraduate and postgraduate courses in various disciplines.",
        "Yes, the college provides bus facilities for students from different locations.",
        "You can access the college Wi-Fi using your student login credentials.",
        "The library is open from 9:00 AM to 5:00 PM on working days.",
        "Yes, students are encouraged to participate in sports, clubs, and cultural activities."
    ]
}
```

```python
faq_df = pd.DataFrame(faq_data)

faq_df
```

```python
# Create TF-IDF Vectorizer
vectorizer = TfidfVectorizer()

# Convert questions into numerical vectors
question_vectors = vectorizer.fit_transform(faq_df["Question"])

print("AI model is ready!")
```

```python
def chatbot(user_question):

    # Convert user question into vector
    user_vector = vectorizer.transform([user_question])

    # Find similarity
    similarity = cosine_similarity(user_vector, question_vectors)

    # Get the most similar question index
    best_match = similarity.argmax()

    # Return corresponding answer
    return faq_df.iloc[best_match]["Answer"]
```

```python
print(chatbot("What are the college timings?"))
```

```python
print(chatbot("How can I pay fees?"))
```

```python
interface = gr.Interface(
    fn=chatbot,
    inputs=gr.Textbox(
        lines=2,
        placeholder="Ask a college-related question..."
    ),
    outputs="text",
    title="🎓 College FAQ Chatbot",
    description="Ask any question related to college."
)
```

```python
interface.launch()
```
