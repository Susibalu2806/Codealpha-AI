```python
languages = {
    "English": "en",
    "Tamil": "ta",
    "Hindi": "hi",
    "Telugu": "te",
    "Malayalam": "ml",
    "Kannada": "kn",
    "French": "fr",
    "German": "de",
    "Spanish": "es",
    "Japanese": "ja",
    "Chinese": "zh-CN"
}

print("Available Languages:")

for language in languages:
    print(language)
```

```python
from deep_translator import GoogleTranslator

def translate_text(text, source, target):

    translated = GoogleTranslator(
        source=languages[source],
        target=languages[target]
    ).translate(text)

    return translated
```

```python
text = input("Enter the text: ")

source = input("Enter Source Language: ")

target = input("Enter Target Language: ")

result = translate_text(text, source, target)

print("\nTranslated Text:")
print(result)
```

```python
import gradio as gr

app = gr.Interface(
    fn=translate_text,
    inputs=[
        gr.Textbox(label="Enter Text"),
        gr.Dropdown(list(languages.keys()), label="Source Language"),
        gr.Dropdown(list(languages.keys()), label="Target Language")
    ],
    outputs=gr.Textbox(label="Translated Text"),
    title="🌍 AI Language Translation Tool",
    description="Translate text instantly between multiple languages using Artificial Intelligence."
)

app.launch(share=True)
```
